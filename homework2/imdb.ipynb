```python
mymovies = mypage.find_all('div', attrs={'class': 'lister-item-content'})
```

```python
mymovie = mymovies[0]
```

```python
mymovie.find()
```

```python
mymovie.find('h3').find('a').text
```

```python
mymovie.find(
    'div', attrs={'class': 'ratings-bar'}
).find(
    'div', attrs={'class': 'inline-block ratings-imdb-rating'}
).find('strong').text
```

```python
movies = []
for mymovie in mymovies:
    mytitle = mymovie.find('h3').find('a').text
    myrating = mymovie.find(
        'div', attrs={'class': 'ratings-bar'}
    ).find(
        'div', attrs={'class': 'inline-block ratings-imdb-rating'}
    ).find('strong').text
    myruntime = mymovie.find(
        'span', attrs={'class': 'runtime'}
    ).text
    movies.append([mytitle, myrating, myruntime])
```

```python
movies
```

```python
import csv
with open('mymovies.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerows(movies)
```
